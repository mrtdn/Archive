# Part2 (ML Modelling)



We are going to use 5 different ML Regression Models to predict Technical Debt values:

>Linear Regression

>Support Vector Regression

>Decision Tree Regression

>Random Forest Tree Regression

>Multi-layer Perceptron

70% of our data will be used as training set and 30% as test set. Training and test sets will be randomly. The “random_state” variable will have a fixed value so that all models get the same random data. In this way, we will be able to evaluate the success rate of the different models objectively.


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.neighbors import NearestCentroid
from sklearn.metrics import r2_score

random_state_fixed=53


#This function plots graphs to improve visual confirmation
def plotGraph(Y_test,Y_pred, regressionName, XVar, YVar):
    plt.figure(figsize=(7, 7))
    plt.scatter(Y_test, Y_test, color = 'red', label = 'Original Values')
    plt.scatter(Y_test, Y_pred, color = 'blue', label = 'Predicted Values')
    plt.title(f'{regressionName} of {YVar} ({XVar})')
    plt.xlabel(f'{XVar}')
    plt.ylabel(f'{YVar} Predictions')
    plt.legend()
    plt.show()
    
    
# This function is an extra one for kernel value.
def plotGraphKernel(Y_test,Y_pred, kernelx, regressionName, XVar, YVar):
    plt.figure(figsize=(6, 6))
    plt.scatter(Y_test, Y_test, color = 'red', label = 'Original Values')
    plt.scatter(Y_test, Y_pred, color = 'blue', label = 'Predicted Values')
    plt.title(f'{regressionName} of {YVar} ({XVar})(Kernel: {kernelx})')
    plt.xlabel(f'{XVar}')
    plt.ylabel(f'{YVar} Predictions')
    plt.legend()
    plt.show()


#Linear Regressor is one of the best choice for the data that is correlated linearly:
def linearRegression(X_train, Y_train, X_test, Y_test, regressionName, XVar, YVar):
    model = LinearRegression().fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    R_square = r2_score(Y_test, Y_pred) 
    print('R-Squared Value:', R_square)
    Mean_square = mean_squared_error(Y_test, Y_pred, squared=False)
    print('Root Mean Squared Error:', Mean_square)
    plotGraph(Y_test,Y_pred, regressionName, XVar, YVar)
    
    

#SVR is also one of the most common regressor that used for regression predictions
def svrRegression(X_train,Y_train,X_test,Y_test, kernelx, regressionName, XVar, YVar):
    model = SVR(kernel = kernelx).fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    R_square = r2_score(Y_test, Y_pred) 
    print('R-Squared Value:', R_square)
    Mean_square = mean_squared_error(Y_test, Y_pred, squared=False)
    print('Root Mean Squared Error:', Mean_square)
    plotGraphKernel(Y_test,Y_pred, kernelx, regressionName, XVar, YVar)


# Decision Tree Regressor
def decisionTreeRegression(X_train,Y_train,X_test,Y_test, regressionName, XVar, YVar, random_state_fixed):
    model = DecisionTreeRegressor(random_state=random_state_fixed).fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    R_square = r2_score(Y_test, Y_pred) 
    print('R-Squared Value:', R_square)
    Mean_square = mean_squared_error(Y_test, Y_pred, squared=False)
    print('Root Mean Squared Error:', Mean_square)
    plotGraph(Y_test,Y_pred, regressionName, XVar, YVar)


# Random Forest Tree Regressor
def randomDecTreeRegression(X_train,Y_train,X_test,Y_test, regressionName, XVar, YVar, random_state_fixed):
    model = RandomForestRegressor(random_state=random_state_fixed).fit(X_train,Y_train)
    Y_pred = model.predict(X_test)
    R_square = r2_score(Y_test, Y_pred) 
    print('R-Squared Value:', R_square)
    Mean_square = mean_squared_error(Y_test, Y_pred, squared=False)
    print('Root Mean Squared Error:', Mean_square)
    plotGraph(Y_test,Y_pred, regressionName, XVar, YVar)



# Multi-layer Perceptron regressor: 'This model optimizes the squared-loss using LBFGS or stochastic gradient descent.'
def mlpRegression(X_train,Y_train,X_test,Y_test, regressionName, XVar, YVar, random_state_fixed):
    model = MLPRegressor(random_state=random_state_fixed).fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    R_square = r2_score(Y_test, Y_pred) 
    print('R-Squared Value:', R_square)
    Mean_square = mean_squared_error(Y_test, Y_pred, squared=False)
    print('Root Mean Squared Error:', Mean_square)
    plotGraph(Y_test,Y_pred, regressionName, XVar, YVar)
    
```

In this step (step 3), we will try to estimate Technical Debt using internal metrics.

```python
# Split data into train and test data

df_normalized_copy = df_normalized.copy()
df_normalized_copy.drop(['NOB','V','SH', 'CS', 'TD', 'CDR'], axis='columns', inplace=True)
X = df_normalized_copy.iloc[:, :].values

df_normalized_copy = df_normalized.copy()
df_normalized_copy.drop(['NOC', 'CBO', 'RFC', 'LOC', 'CLOC', 'FanIn', 'FanOut', 'LCOM', 'WMC', 'DIT', 'NOB','V','SH', 'CS', 'TD'], axis='columns', inplace=True)
YCDR = df_normalized_copy.iloc[:, :].values

from sklearn.model_selection import train_test_split
X_train1, X_test1, Y_train1, Y_test1 = train_test_split(X, YCDR, test_size = 0.30, random_state = random_state_fixed)

df_normalized_copy = df_normalized.copy()
df_normalized_copy.drop(['NOC', 'CBO', 'RFC', 'LOC', 'CLOC', 'FanIn', 'FanOut', 'LCOM', 'WMC', 'DIT', 'NOB','V','SH', 'CS', 'CDR'], axis='columns', inplace=True)
YTD = df_normalized_copy.iloc[:, :].values

X_train2, X_test2, Y_train2, Y_test2 = train_test_split(X, YTD, test_size = 0.30, random_state = random_state_fixed)
```

```python
#Model for Linear Regression (show also R-Squared Value and Root Mean Squared Error )
linearRegression(X_train1,Y_train1,X_test1,Y_test1, 'Linear Regression', 'Internal Metrics', 'CDR')
print('-----------------------------------------------------------------\n')
linearRegression(X_train2,Y_train2,X_test2,Y_test2, 'Linear Regression', 'Internal Metrics', 'TD')
```

```python
#Model for SVR Regression (show also R-Squared Value and Root Mean Squared Error )
svrRegression(X_train1,Y_train1.ravel(),X_test1,Y_test1.ravel(), 'linear', 'SVR Regression', 'Internal Metrics', 'CDR')
print('-----------------------------------------------------------------\n')
svrRegression(X_train2,Y_train2.ravel(),X_test2,Y_test2.ravel(), 'linear', 'SVR Regression', 'Internal Metrics', 'TD')
print('-----------------------------------------------------------------\n')
svrRegression(X_train1,Y_train1.ravel(),X_test1,Y_test1.ravel(), 'rbf', 'SVR Regression', 'Internal Metrics', 'CDR')
print('-----------------------------------------------------------------\n')
svrRegression(X_train2,Y_train2.ravel(),X_test2,Y_test2.ravel(), 'rbf', 'SVR Regression', 'Internal Metrics', 'TD')
```

```python
#Model for Decision Tree Regression (show also R-Squared Value and Root Mean Squared Error )
decisionTreeRegression(X_train1,Y_train1,X_test1,Y_test1, 'Decision Tree Regression', 'Internal Metrics', 'CDR', random_state_fixed)
print('-----------------------------------------------------------------\n')
decisionTreeRegression(X_train2,Y_train2,X_test2,Y_test2, 'Decision Tree Regression', 'Internal Metrics', 'TD', random_state_fixed)
```

```python
#Model for Random Forest Tree Regression (show also R-Squared Value and Root Mean Squared Error )
randomDecTreeRegression(X_train1,Y_train1.ravel(),X_test1,Y_test1.ravel(), 'Random Forest Tree Regression', 'Internal Metrics', 'CDR', random_state_fixed)
print('-----------------------------------------------------------------\n')
randomDecTreeRegression(X_train2,Y_train2.ravel(),X_test2,Y_test2.ravel(), 'Random Forest Tree Regression', 'Internal Metrics', 'TD', random_state_fixed)
```

```python
#Model for Multi-layer Perceptron Regressor (show also R-Squared Value and Root Mean Squared Error )
mlpRegression(X_train1,Y_train1.ravel(),X_test1,Y_test1.ravel(), 'MultiLayer Perceptron Regression', 'Internal Metrics', 'CDR', random_state_fixed)
print('-----------------------------------------------------------------\n')
mlpRegression(X_train2,Y_train2.ravel(),X_test2,Y_test2.ravel(), 'MultiLayer Perceptron Regression', 'Internal Metrics', 'TD', random_state_fixed)
```
